```python
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.impute import SimpleImputer
```

```python
def biuld_preprocessor(x:pd.DataFrame):
    numreic_features = x.select_dtypes(include=['int64','float64']).columns.tolist()
    categorical_features = x.select_dtypes(include=['object']).columns.tolist()
    
    numeric_transformer = Pipeline(steps=[
        ('imputer',SimpleImputer(strategy='median'),
        'scaling',StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ecoder',OneHotEncoder(handle_unkown = 'ignore'))
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num',numreic_features,numeric_transformer),
            ('cat',categorical_features,categorical_transformer)
        ]
    )

    return preprocessor
```
